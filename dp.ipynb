## Dynamic Programming 
Solving smaller problems to then use those smaller solutions to buyild a solution to the larger problem.


```python
#O(n) time | o(n) space
def maxSubsetSumNoAdjacent(array):
    # Write your code here.
    #build array of same length --> maxSums
    
    #formula: maxSums[i] = max(maxSums[i-1], maxSums[i-2] + array[i])
    #basically using the previous 
    if not len(array):
        return 0
    elif len(array) == 1:
        return array[0] 

    #base cases
    maxSums = array[:] #copy main array 
    #maxsum[0] = array[0] 
    maxSums[1] = max(array[0] , array[1])

    #start from 2 bacuase 0 and 1 are taken care of already 
    for i in range(2,(len(array))):
        maxSums[i] = max(maxSums[i-1], maxSums[i-2] + array[i])
    return(maxSums[-1])#return last value 
```
